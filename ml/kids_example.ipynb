```python
pd.set_option('display.max_rows', 200)
df_kids = (df.set_index('order_id').loc[ kids_orders ]).reset_index(drop=False)
df_kids.head(200)
```

```python
df_entrees_kids_ids = list(set(list(df_kids[ df_kids['item_category_type'] == 'Entree']['order_id']
                                   )))

print(len(kids_orders))
print(len(df_entrees_kids_ids))

exc_kids_ids = list(set(kids_orders)^set(df_entrees_kids_ids))

print(len(exc_kids_ids))

df_only_kids = (df.set_index('order_id').loc[ exc_kids_ids ]).reset_index(drop=False)
df_only_kids.head(100)
```

```python
xdf = pd.DataFrame(columns=df_only_kids.columns)

for i in range(len(exc_kids_ids)):
    
    dd = df_only_kids.loc[ df_only_kids['order_id'] == exc_kids_ids[i]]
    
    ndessert = len(dd[ dd['item_category_type'] == 'Dessert'])
    nbev = len(dd[ dd['item_category_type'] == 'Beverage'])
    nside = len(dd[ dd['item_category_type'] == 'Side'])
    nalc = len(dd[ dd['item_category_type'] == 'Alcoholic Beverage'])
    
    if nbev == 0 and nside == 0 and ndessert == 0:
        continue
    elif nalc > 0:
        continue
    else:
        xdf = xdf.append(dd)
    

xdf.head(200)
```
